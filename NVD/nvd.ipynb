```python
import requests
import json
import time
import pandas as pd

headers = {'Content-Type': 'application/json'}
pagesize=100
total_cves=1000
startindex=0
cve_array = []

while (startindex < total_cves):
    response=requests.get("https://services.nvd.nist.gov/rest/json/cves/2.0/?resultsPerPage={}&startIndex={}".format(pagesize, startindex),headers=headers)
    cves=json.loads(response.content.decode('utf-8'))
    for i in cves['vulnerabilities']:
        cve_array.append(i)
    time.sleep(10)
    startindex=startindex+pagesize
 
nvd_data = pd.json_normalize(cve_array)
nvd_data
```
